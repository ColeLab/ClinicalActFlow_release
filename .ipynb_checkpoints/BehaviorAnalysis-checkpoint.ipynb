# SCAP behaviour script
- Overall accuracy and RT
- Mixed Anova
- Correlations with other behaviour

```python
import numpy as np
import pandas as pd
from functions.data_helpers import load_scan_behaviour
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg

#subject list
subj_df = pd.read_csv('subject_list.txt', sep='\t', index_col = 0, header = 0)
figure_dir = '/projects/f_mc1689_1/ClinicalActFlow/docs/figures/N=93/'
PAL = sns.xkcd_palette(['grey','light orange'])
```

## Load data

```python
accuracy,reaction_time,_,_,task_labels = load_scan_behaviour(subj_df.index,['scap'],subj_df['group'].values)
```

# Overall accuracy and reaction time

```python
# do a t-test on total ACC
x = np.mean(np.mean(accuracy['scap']['CTRL'],axis=0),axis=0) * 100
y = np.mean(np.mean(accuracy['scap']['SCHZ'],axis=0),axis=0) * 100
display(pg.ttest(x,y,paired=False))
print('ACC: CTRL mean:',np.round(np.mean(x),2),'SCHZ mean:',np.round(np.mean(y),2))

# put arrays into dataframe for seaborn
plot_df = pd.DataFrame()
plot_df['ACC'] = np.hstack((x,y))

# # do a t-test on mean RT on correct trials (hence nanmean)
x = np.nanmean(np.nanmean(reaction_time['scap']['CTRL'],axis=0),axis=0)
y = np.nanmean(np.nanmean(reaction_time['scap']['SCHZ'],axis=0),axis=0)
display(pg.ttest(x,y,paired=False))
print('RT: CTRL mean:',np.round(np.mean(x)*1000,2),'SCHZ mean:',np.round(np.mean(y)*1000,2))

# save out in df for plotting
plot_df['RT'] = np.hstack((x,y))
plot_df['group'] = np.hstack(((np.ones(x.shape)-1),np.ones(y.shape)))
```

```python
# create figure
plt.figure(figsize=(4,2))

# do plots
plt.subplot(121)
sns.swarmplot(x='group', y='ACC',data=plot_df,size=3, color=".1", linewidth=0,alpha=0.65)
ax = sns.boxplot(x='group', y='ACC',data=plot_df,whis=0,fliersize=0)

# Tweak the visual presentation
ax.set(ylabel='Accuracy (%)')
ax.set(xlabel='')
ax.set_xticklabels(['HC','SC'])
sns.despine(trim=True)

plt.subplot(122)
sns.swarmplot(x='group', y='RT',data=plot_df,size=3, color=".1", linewidth=0,alpha=0.65)
ax = sns.boxplot(x='group', y='RT',data=plot_df,whis=0,fliersize=0)

# Tweak the visual presentation
ax.set(ylabel='Reaction time (s)')
ax.set(xlabel='')
ax.set_xticklabels(['HC','SC'])
sns.despine(trim=True)

plt.tight_layout()
plt.savefig(figure_dir+'behav-overall.svg',dpi=600)
plt.savefig(figure_dir+'behav-overall.jpeg',dpi=600)
plt.show()
```

# Mixed ANOVA
Main effect of working memory and group with no interaction

```python
# get the data
x = np.vstack((np.mean(np.mean(accuracy['scap']['CTRL'][:,0:6,:],axis=0),axis=0) * 100,
                  np.mean(np.mean(accuracy['scap']['CTRL'][:,6::,:],axis=0),axis=0) * 100)).T
y = np.vstack((np.mean(np.mean(accuracy['scap']['SCHZ'][:,0:6,:],axis=0),axis=0) * 100,
                  np.mean(np.mean(accuracy['scap']['SCHZ'][:,6::,:],axis=0),axis=0) * 100)).T

# create mixed ANOVA dataframe
df_anova = pd.DataFrame()
n1 = len(x)
n2 = len(y)
n_tot = n1 + n2

# group status
group_vec = np.hstack((['HC']*n1,['SZ']*n2))
df_anova['Group'] = np.hstack((group_vec,group_vec))

# subj number
df_anova['subj'] = np.hstack((np.arange(1,n_tot+1),
                              np.arange(1,n_tot+1)))

df_anova['wm_load'] = np.hstack((np.ones(n_tot),np.ones(n_tot)+1))

# accuracy
df_anova['acc'] = np.hstack((x[:,0],y[:,0],x[:,1],y[:,1]))

# get reaction time data
# get the data
x = np.vstack((np.nanmean(np.nanmean(reaction_time['scap']['CTRL'][:,0:6,:],axis=0),axis=0),
                  np.nanmean(np.nanmean(reaction_time['scap']['CTRL'][:,6::,:],axis=0),axis=0))).T
y = np.vstack((np.nanmean(np.nanmean(reaction_time['scap']['SCHZ'][:,0:6,:],axis=0),axis=0),
                  np.nanmean(np.nanmean(reaction_time['scap']['SCHZ'][:,6::,:],axis=0),axis=0))).T
df_anova['rt'] = np.hstack((x[:,0],y[:,0],x[:,1],y[:,1]))

# do the anova
aov = pg.mixed_anova(dv='acc', between='Group', within='wm_load', subject='subj', data=df_anova)
display(aov)
print('p-values (without notation)',np.round(aov['p-unc'].values,3))
aov = pg.mixed_anova(dv='rt', between='Group', within='wm_load', subject='subj', data=df_anova)
display(aov)
print('p-values (without notation)',np.round(aov['p-unc'].values,3))
```

```python
# create figure
plt.figure(figsize=(3*.90,2*.90))

# do plots
ax1 = sns.pointplot(x='wm_load', y='acc', hue='Group',data=df_anova,kind="point",
                    dodge=True,scale=0.75,capsize=.2,linewidth=1.5,errwidth=1.5,palette=PAL)

# Tweak the visual presentation
ax1.set(ylabel='Accuracy (%)')
ax1.set(xlabel='Working memory load')
ax1.set_xticklabels(['Low','High'])
sns.despine(trim=True)
ax1.legend(frameon=False)
plt.savefig(figure_dir+'behav-acc-ANOVA.svg',dpi=600,bbox_inches='tight',pad_inches=1)

# do plots
plt.figure(figsize=(3,2))
ax1 = sns.pointplot(x='wm_load', y='rt', hue='Group',data=df_anova,kind="point",
                    dodge=True,scale=0.75,capsize=.2,linewidth=1.5,errwidth=1.5,palette=PAL)

# Tweak the visual presentation
ax1.set(ylabel='Reaction time (s)')
ax1.set(xlabel=None)
ax1.set_xticklabels(['Low','High'])
sns.despine(trim=True)
ax1.legend(frameon=False)

#plt.tight_layout()
#plt.savefig(figure_dir+'behav-ANOVA.jpeg',dpi=600,bbox_inches='tight',pad_inches=1)
plt.show()
```

# Correlations

```python
from scipy.stats import spearmanr
from functions.data_helpers import beh_tsv_helper,load_ind_diff_behaviour

# get the data
_,beh_tsv_dict,_ = beh_tsv_helper()
beh_data,beh_labels = load_ind_diff_behaviour(subj_df.index, beh_tsv_dict)

# sort data into relevant groups for a correlation
x = np.hstack((beh_data[:,subj_df['group']=='CTRL'],beh_data[:,subj_df['group']=='SCHZ'])).T

# collate the scap task
scap_acc = np.hstack((np.mean(np.mean(accuracy['scap']['CTRL'],axis=0),axis=0) * 100,np.mean(np.mean(accuracy['scap']['SCHZ'],axis=0),axis=0) * 100))
scap_rt = np.hstack((np.nanmean(np.nanmean(reaction_time['scap']['CTRL'],axis=0),axis=0),np.nanmean(np.nanmean(reaction_time['scap']['SCHZ'],axis=0),axis=0)))

# add variable to array and flip (so that key variable is first)
y = np.fliplr(np.hstack((x,scap_acc.reshape(len(scap_acc),-1))))

# choose whether to use rank or regular correlation
#corr_mat = np.corrcoef(y.T)
corr_mat,p_mat = spearmanr(y)
corr_vec = corr_mat[0,1::] # the correlations associated with scap_acc
p_vec = p_mat[0,1::]

# put in a dataframe and plot
plot_df = pd.DataFrame()
plot_df['Measure'] = beh_labels
plot_df['r'] = corr_vec
plot_df['p'] = p_vec
plot_df['p_bonf'] = p_vec * len(p_vec)

plt.figure(figsize=(4,10))
ax = sns.barplot(x='r',y='Measure',data=plot_df.sort_values('r', ascending=False))
# Tweak the visual presentation
ax.set(ylabel='')
ax.set(xlabel='Correlation with SCAP accuracy')
sns.despine(trim=True)

plt.savefig(figure_dir+'behav-individual_differences.svg',dpi=600)
plt.savefig(figure_dir+'behav-individual_differences.jpeg',dpi=600)
plt.show()
```
